```python
data = np.loadtxt('/content/gdrive/MyDrive/Machine Learning/q2-data/Compound.txt')
data_purity = data.copy()
data[:,-1] = 0
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c='darkcyan')
plt.show()
db = DBSCAN(eps=1.6, min_samples=12).fit(data)
core_samples = db.core_sample_indices_
labels = db.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
my_polet = 10*["olivedrab",'chocolate','c','crimson','steelblue','orangered','mediumpurple','goldenrod','indigo','coral','powderblue','firebrick','yellowgreen','orchid','skyblue','pink','springgreen','maroon','r','g','b','y']
```

```python
def plot_result(labels,core_samples,n_clusters_,data):
  pl.close('all')
  pl.figure(1,figsize=(10,7))
  pl.clf()

  # Black removed and is used for noise instead.
  colors = my_polet
  for k, col in zip(set(labels), colors):
      if k == -1:
          # Black used for noise.
          col = 'k'
          markersize = 3
      
      class_members = [index[0] for index in np.argwhere(labels == k)]
      cluster_core_samples = [index for index in core_samples
                              if labels[index] == k]
      for index in class_members:
          x = data[index]
          if index in core_samples and k != -1:
              markersize = 10
          else:
              markersize = 9
          pl.plot(x[0], x[1], 'o', markerfacecolor=col,
                  markeredgecolor='k', markersize=markersize)

  pl.title('Estimated number of clusters: %d' % n_clusters_)
  pl.show() 
```

```python
plot_result(labels,core_samples,n_clusters_,data)
```

```python
##calculate Purity
purity = 0.0
Real_number_of_clusters = 6
for i in range(Real_number_of_clusters):
    temp = np.zeros(len(labels))
    for j in range(len(labels)):
        if int(data_purity[j][-1]) == i+1 and labels[j] != -1:
            temp[int(labels[j])] += 1
    purity += (1/len(labels))*np.max(temp)
print('Purity',purity)
```
