```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data = pd.read_excel('/content/gdrive/MyDrive/Machine Learning/q3-data/data_h.xlsx').values
plt.scatter(data[:,0],data[:,1],c='c')
plt.show()
```

```python
clusterings = []
for i in range(2,9):
  clustering = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='single')
  clustering.fit_predict(data)
  clusterings.append(clustering)
```

```python
f, axes = plt.subplots(nrows = 4, ncols = 2,figsize=(15,20))
index = 0 
for i in range (4):
  for j in range(2):
    if index <7:
      axes[i][j].scatter(data[:,0], data[:,1], c=clusterings[index].labels_, cmap='rainbow')
      axes[i][j].set_title("number of clusters: "+str(index+2))
      index+=1
plt.savefig('single_topDown.png')
```

```python
def topDown(distance_measure):
  clusterings = []
  for i in range(2,9):
    clustering = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage=distance_measure)
    clustering.fit_predict(data)
    clusterings.append(clustering)
  f, axes = plt.subplots(nrows = 4, ncols = 2,figsize=(15,20))
  index = 0 
  for i in range (4):
    for j in range(2):
      if index <7:
        axes[i][j].scatter(data[:,0], data[:,1], c=clusterings[index].labels_, cmap='rainbow')
        axes[i][j].set_title("number of clusters: "+str(index+2))
        index+=1
  f.suptitle('Top Down clustering - distance measure : '+ distance_measure +" link",fontsize=20)
  plt.savefig(distance_measure + '_topDown.png')
```

```python
topDown("single")
topDown("average")
topDown("complete")
```

```python
def bottomUp(distance_measure):
  clusterings = []
  for i in range(2,9):
    clustering = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage=distance_measure)
    clustering.fit_predict(data)
    clusterings.append(clustering)
  f, axes = plt.subplots(nrows = 4, ncols = 2,figsize=(15,20))
  index = 6
  for i in range (4):
    for j in range(2):
      if index >-1:
        axes[i][j].scatter(data[:,0], data[:,1], c=clusterings[index].labels_, cmap='rainbow')
        axes[i][j].set_title("number of clusters: "+str(index+2))
        index-=1
  f.suptitle('Bottom Up clustering - distance measure : '+ distance_measure +" link",fontsize=20)
  plt.savefig(distance_measure + '_bottomUp.png')
```

```python
bottomUp("single")
bottomUp("average")
bottomUp("complete")
```
